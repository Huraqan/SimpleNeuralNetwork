# Single epoch comparison

```python
import pandas as pd
import seaborn as sns
from notebook_tools import side_by_side, get_win_fails, get_win_fails_by_digit, plot_win_fails

costs_normal_16_16 = pd.read_csv("reports/costs_normal_16_16.csv")
costs_normal_49_16 = pd.read_csv("reports/costs_normal_49_16.csv")
costs_noise_16_16 = pd.read_csv("reports/costs_noise_16_16.csv")
costs_noise_49_16 = pd.read_csv("reports/costs_noise_49_16.csv")
```

```python
side_by_side({
    "Normal 16-16": get_win_fails(costs_normal_16_16),
    "Normal 49-16": get_win_fails(costs_normal_49_16),
    "Noise 16-16": get_win_fails(costs_noise_16_16),
    "Noise 49-16": get_win_fails(costs_noise_49_16),
})
```

```python
plot_win_fails(costs_normal_16_16)
```

```python
plot_win_fails(costs_normal_49_16)
```

```python
plot_win_fails(costs_noise_16_16)
```

```python
plot_win_fails(costs_noise_49_16)
```

```python
side_by_side({
    "Normal 16-16": get_win_fails_by_digit(costs_normal_16_16),
    "Normal 49-16": get_win_fails_by_digit(costs_normal_49_16),
    "Noise 16-16": get_win_fails_by_digit(costs_noise_16_16),
    "Noise 49-16": get_win_fails_by_digit(costs_noise_49_16),
})
```

```python
sns.histplot(x=costs_normal_16_16["Digit"].astype(str), y=costs_normal_16_16["Cost"], legend=False).set_facecolor("k")
```

```python
sns.histplot(x=costs_normal_49_16["Digit"].astype(str), y=costs_normal_49_16["Cost"]).set_facecolor("k")
```

```python
sns.histplot(x=costs_noise_16_16["Digit"], y=costs_noise_16_16["Cost"]).set_facecolor("k")
```

```python
sns.histplot(x=costs_noise_49_16["Digit"], y=costs_noise_49_16["Cost"]).set_facecolor("k")
```
